# Pandas Data Visualization

```python
# Import libraries
import numpy as np  
import pandas as pd
import seaborn as sns
%matplotlib inline
```

**# Read Data**  
df1 = pd.read_csv('df1', index_col = 0)  
df2 = pd.read_csv('df2')  

**# Create histogram**    
df1['A'].hist(bins = 30)  
df1['A'].plot(kind = 'hist', bins = 30)  
df1['A'].plot.hist(bins = 50)  

**# Create area plot**  
df2.plot.area(alpha = 0.4)

**# Create bar plot**  
df2.plot.bar(stacked = True)

**# Create line plot**  
df1.plot.line(x = df1.index, y = 'B', figsize = (12,3), lw = 1)

**# Create scatter plot using color**  
df1.plot.scatter(x = 'A', y = 'B', c = 'C, cmap = 'coolwarm')

**# Create scatter plot using size**  
df1.plot.scatter(x = 'A', y = 'B', s = df1['C']*100)

**# Create box plot**  
df2.plot.box()

### Bivariate Data

```python
# Create data frame
df = pd.DataFrame(np.random.randn(1000,2), columns = ['a','b'])
df.head()
```

```python
# Create hex bin
df.plot.hexbin(x = 'a',y = 'b', gridsize = 25, cmap = 'coolwarm')
```

```python
# Create kernal density estimation plot
df['a'].plot.kde()
```

```python
# Create kernal density estimation plot
df.plot.density()
```
